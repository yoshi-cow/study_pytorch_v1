```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
```

## data load

```python
data_path = '/content/drive/MyDrive/study_DeepLearning/data/AMZN.csv'
data = pd.read_csv(data_path)
data
```

```python
data = data[['Date', 'Close']]
data
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
device
```

```python
data['Date'] = pd.to_datetime(data['Date'])

plt.plot(data['Date'], data['Close'])
plt.show
```

```python
# lag列設定
# sequence length列の設定
# 本来は、Dataset classの __getitem__メソッド内でsequence lengthの数づつ取り出すが
# 読み取る時系列の順番を逆にしたいため、ここで作っている？
# ならば、__getitem__メソッド内で順番を逆にするコードを入れればいいと思うが・・・
from copy import deepcopy as dc

def prepare_dataframe_for_lstm(df, n_steps):
  df = dc(df)
  df['Date'] = pd.to_datetime(df['Date'])

  df.set_index('Date', inplace=True)

  for i in range(1, n_steps + 1):
    df[f'Close(t-{i})'] = df['Close'].shift(i)

  df.dropna(inplace=True)

  return df

lookback = 7
shifted_df = prepare_dataframe_for_lstm(data, lookback)
shifted_df
```

```python
shifted_df_as_np = shifted_df.to_numpy()

shifted_df_as_np
```

## normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1, 1))
shifted_df_as_np = scaler.fit_transform(shifted_df_as_np)

shifted_df_as_np
```

```python
# データをx, yに分割
X = shifted_df_as_np[:, 1:]
y = shifted_df_as_np[:, 0]

X.shape, y.shape
```

## LSTMが古いデータから読み取るように、xの列順番を古い順(t-7, t-6, t-5,...)に変更

```python
X = dc(np.flip(X, axis=1))
X
```

## train test split
* 95%を学習に、残り5%をテストに使用する

```python
# 学習とテストを分割するindex位置算出
split_index = int(len(X) * 0.95)

split_index
```

```python
X_train = X[:split_index]
X_test = X[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Dataset, DataLoader作成

```python
# pytorchにあてはめられるように各データに次元を追加する
# この作業の意味がよくわからない？？？？？
X_train = X_train.reshape((-1, lookback, 1))
X_test = X_test.reshape((-1, lookback, 1))

y_train = y_train.reshape((-1, 1))
y_test = y_test.reshape((-1, 1))

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# tensorに変換
X_train = torch.tensor(X_train).float()
y_train = torch.tensor(y_train).float()
X_test = torch.tensor(X_test).float()
y_test = torch.tensor(y_test).float()

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Datasetクラスの定義
from torch.utils.data import Dataset

class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.y[i]

train_dataset = TimeSeriesDataset(X_train, y_train)
test_dataset = TimeSeriesDataset(X_test, y_test)
```

```python
# DataLoaderの作成（バッチ処理用）
from torch.utils.data import DataLoader

batch_size = 16

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# 1バッチの中身確認
for _, batch in enumerate(train_loader):
    x_batch, y_batch = batch[0].to(device), batch[1].to(device)
    print(x_batch.shape, y_batch.shape)
    break
```

## Modeling

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_stacked_layers):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_stacked_layers = num_stacked_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_stacked_layers,
                            batch_first=True) # batch_first=Trueで、入力データの最初の次元がバッチサイズの次元であることを示す

        self.fc = nn.Linear(hidden_size, 1) # 出力値は1要素なので、最終layerの出力値は1つ

    def forward(self, x):
        batch_size = x.size(0)
        # h0とc0ゲートの初期値設定(torch.zeros(引数)で、引数の次元で全要素が0のtensorができる)
        h0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size).to(device)
        c0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size).to(device)

        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

model = LSTM(1, 4, 1)
model.to(device)
model
```

```python
# エポックごとの処理を関数にまとめる
# 通常は、エポックのループの中に、バッチサイズずつ処理をループするを書くが、
# 関数化により、ループがネストされたコードを避けることで、見やすいコードになる
def train_one_epoch():
  model.train(True)
  print(f'Epoch {epoch+1}')
  running_loss = 0.0

  for batch_index, batch in enumerate(train_loader):
    x_batch, y_batch = batch[0].to(device), batch[1].to(device)

    output = model(x_batch)
    loss = loss_function(output, y_batch)
    running_loss += loss.item()

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if batch_index % 100 == 99: # print every 100 batches
      avg_loss_across_batches = running_loss / 100
      print(f'Batch {batch_index+1} loss: {avg_loss_across_batches:.3f}')
      running_loss = 0.0
  print()
```

```python
def validate_one_epoch():
  model.train(False)
  running_loss = 0.

  for batch_index, batch in enumerate(test_loader):
    x_batch, y_batch = batch[0].to(device), batch[1].to(device)

    with torch.no_grad():
      output = model(x_batch)
      loss = loss_function(output, y_batch)
      running_loss += loss.item()

  avg_loss_across_batches = running_loss / len(test_loader)

  print(f'Val Loss: {avg_loss_across_batches:.3f}')
  print('***************************************************')
  print()
```

```python
learning_rate = 0.001
num_epochs = 10
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    train_one_epoch()
    validate_one_epoch()
```

```python
# 学習時の予測結果可視化
with torch.no_grad():
    predicted = model(X_train.to(device)).to('cpu').numpy()

plt.plot(y_train, label='Actual Close')
plt.plot(predicted, label='Predicted Close')
plt.xlabel('Day')
plt.ylabel('Close')
plt.legend()
plt.show()
```
