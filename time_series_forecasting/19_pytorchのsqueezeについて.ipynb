# Tensor.squeeze()について
* Tensor.squeeze()は「テンソル中の要素数1の次元を削除する」
* Tensor.unsqueeze(d: int)は「テンソルのd階目に要素数1の次元を挿入する」 というものだ。

```python
import torch
```

# Unsqueeze使用例
* Tensor.unsqueeze(d: int) - テンソルのd階(次元)目に要素数1の次元を挿入する

```python

t = torch.Tensor([[1, 2, 3], [4, 5, 6]])
t
```

```python
t.shape
```

```python
# 次元数
t.dim()
```

```python
# 0番目の位置(toch.Size([2,3])の2の前の次元)に次元を追加する
t_us = t.unsqueeze(0)
t_us.shape
```

```python
t_us.dim()
```

* ↑ 新しく、もう1次元が増えて3次元になった

```python
# 要素数
t_us.numel()
```

```python
# 1番目の位置(行)に次元を追加する
t_us_1 = t.unsqueeze(1)
display(t_us_1)
print('\n')
print(t_us_1.shape)
```

```python
# 次元数
t_us_1.dim()
```

```python
# 要素数
t_us_1.numel()
```

```python
# 2番目の位置(列)に次元を追加する
t_us_2 = t.unsqueeze(2)
display(t_us_2)
print('\n')
print(t_us_2.shape)
```

```python
# 次元数
t_us_2.dim()
```

# Squeeze
* テンソル内の要素数1の次元を削除する

```python
t
```

```python
t.squeeze()
```

 * ↑ 要素数が1の次元がないので変化なし

```python
display(t_us_1)
print('\n', t_us_1.shape)
```

```python
t_s_1 = t_us_1.squeeze()
display(t_s_1)
print('\n',t_s_1.shape)
```

* ↑ 要素数が1の次元が削除された

```python
display(t_us_2)
print('\n', t_us_2.shape)
```

```python
t_s_2 = t_us_2.squeeze()
display(t_s_2)
print('\n',t_s_2.shape)
```

* ↑ 要素数が1の次元が削除された

```python
display(t_us_2)
print('\n', t_us_2.shape)
```

```python
t_s_2_1 = t_us_2.squeeze(-1) # -1で最後の次元の要素数が1なら最後の次元を削除する！！！
display(t_s_2_1)
print('\n',t_s_2_1.shape)
```

* squeeze(-1)で、最後の次元の要素数が1つだけなら、最後の次元を削除する
