# Normalization method of divi by first record
when dealing with time series data such as stock prices, various methods of normalization can be used, each with its own purpose and interpretation.

## Division by First Day's Price:
This type of normalization, often referred to as `indexing` in finance, is used to compare the performance of different assets over time. By dividing all the prices by the first day's price, you effectively set the initial value of each asset to 1 (or 100 if you multiply by 100 to get a percentage) on the first day. This allows you to see the relative growth or decline of each asset as a straightforward percentage change from the initial value. It's particularly useful for visualizing and comparing the performance of multiple assets on the same chart.

**Pros**:
* Easy to interpret, as it shows the percentage change from the start date.
* Does not assume any distribution of data, making it non-parametric.
* Retains the true temporal structure and magnitude of growth or decline.


**Cons**:
* Sensitive to the choice of the start date.
* Does not adjust for volatility or differing ranges of the data.


## MinMaxScaler:
**MnMaxScaler** is a common feature scaling method that shinks the range of the data to a defined range, typically [0, 1]. The formula for MinMaxScaler is given by:

$$
\text{Normalized Value} = \frac{\text{Value} - \text{Min Value}}{\text{Max Value} - \text{Min Value}}
$$

In finance, this type of scaling could be used when you want to neutralize the effect of the absolute magnitude of prices and bring multiple assets to the same scale so that they can be compared or inputted into models on an even footing.


**Pros**:
* Bounds the data within a specific range, which is helpful for certain algorithms that are sensitive to the scale of data.
* Maintains the distribution of the data.


**Cons**:
* Can be sensitive to outliers since the min and max values can skew the scaling.
* Does not retain the actual magnitude of changes in the data, which might be important for interpreting financial returns.


## Key Differences:
1. **Interpretation**: Dividing by the first day's price indexes the price series, making it easy to interpret grouwth. MinMax scaling transforms data into a bouded interval, which is useful for modeling but less intuitive for performance interpretation.
2. **Sensitivity to Extreames**: the division by the first day's price is unaffected by outliers, whereas MinMax scaling is highly sensitive to the minimum and maximum values.
3. **Purpose of Use**: Division by the first day's price is mainly used for comparative performance visualization, while MinMax scaling is used to prepare features for machine learning algorithms that require data normalization.


In portfolio analysis and prediction, the choice of normalizaiton technique depends on the objective. If the goal is to visualize and compare asset performance over time, indexing by the first day's price is straightforward and effective. If the goal is to input the data into a predictive model, then MinMax scaling or other feature scaling methods might be more appropriate to ensure that the scale of the data doesn't unduly(不当に) influence the model.
