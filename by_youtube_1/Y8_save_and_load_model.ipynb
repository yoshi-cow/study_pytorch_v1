### example

```python
class Model(nn.Module):
  def __init__(self, n_input_features):
    super(Model, self).__init__()
    self.linear = nn.Linear(n_input_features, 1)

  def forward(self, x):
    y_pred = torch.sigmoid(self.linear(x))
    return y_pred

model = Model(n_input_features=6)
```

```python
# save by lazy version
FILE = '/content/drive/MyDrive/study_DeepLearning/data/y8_model.pth'
torch.save(model, FILE)
```

```python
# load
model = torch.load(FILE)
model.eval()

for param in model.parameters():
  print(param)
```

```python
# save by another way
torch.save(model.state_dict(), FILE)
```

```python
# load
loaded_model = Model(n_input_features=6)
loaded_model.load_state_dict(torch.load(FILE))
loaded_model.eval()

for param in loaded_model.parameters():
  print(param)
```

```python
print(loaded_model.state_dict())
```

## 学習過程の各チェックポイントも保存する方法

```python
model = Model(n_input_features=6)

learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
print(optimizer.state_dict())
```

```python
checkpoint = {
    "epoch": 90,
    "model_state": model.state_dict(),
    "optim_state": optimizer.state_dict()
}
checkpoint_FILE = '/content/drive/MyDrive/study_DeepLearning/data/checkpoint.pth'
torch.save(checkpoint, checkpoint_FILE)
```

```python
# checkpointのロード
loaded_checkpoint = torch.load(checkpoint_FILE)
epoch = loaded_checkpoint["epoch"]
```

```python
model = Model(n_input_features=6)
optimizer = torch.optim.SGD(model.parameters(), lr=0)

model.load_state_dict(checkpoint["model_state"])
optimizer.load_state_dict(checkpoint["optim_state"])
```

```python
print(optimizer.state_dict())
```

## GPUの場合

### Save on GPU, Load on CPU

```python
# save on gpu, load on cpu
device = torch.device("cuda")
model.to(device)
torch.save(model.state_dict(), PATH)

device = torch.device("cpu")
model = Model(*args, **kwargs)
model.load_state_dict(torch.load(PATH, map_location=device))
```

### Save on GPU, Load on GPU

```python
# save on gpu, load on gpu
device = torch.device("cuda")
model.to(device)
torch.save(model.satate_dict(), PATH)

model = Model(*args, *:kwargs)
model.load_state_dict(torch.load(PATH))
model.to(device)
```
